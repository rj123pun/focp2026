```python
# Task 1 
name = input("Enter your name: ")

# Checking if the user entered a name
if name.strip() == "":
    print("Hello, Stranger!")
else:
    print(f"Hello, {name}!")
```

```python
# Task 2 password confirmation program
password1 = input("Enter your new password: ")
password2 = input("Confirm your new password: ")

# Checking if two passwords match eachother
if password1 == password2:
    print("Password Set")
else:
    print("Error: Passwords do not match!")
```

```python
# Task 3 Password must be between 8 and 12 characters 
password1 = input("Enter your new password (8-12 characters): ")
password2 = input("Confirm your new password: ")

# Checking password length and match the password to set
if password1 != password2:
    print("Error: Passwords do not match!")
elif len(password1) < 8 or len(password1) > 12:
    print("Error: Password must be between 8 and 12 characters.")
else:
    print("Password Set Successfully")
```

```python
# Task 4: Check against common passwords to set strong password
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

password1 = input("Enter your new password: ")
password2 = input("Confirm your new password: ")

if password1 != password2:
    print("Error: Passwords do not match!")
elif len(password1) < 8 or len(password1) > 12:
    print("Error: Password must be between 8 and 12 characters.")
elif password1.lower() in BAD_PASSWORDS:
    print("Error: Password is too common, choose a stronger one.")
else:
    print("Password Set Successfully")
```

```python
# Task 5 Keep asking until the user sets a valid password
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

while True:
    password1 = input("Enter your new password: ")
    password2 = input("Confirm your new password: ")

    if password1 != password2:
        print("Error: Passwords do not match! Try again.\n")
    elif len(password1) < 8 or len(password1) > 12:
        print("Error: Password must be between 8 and 12 characters. Try again.\n")
    elif password1.lower() in BAD_PASSWORDS:
        print("Error: Password is too common. Try again.\n")
    else:
        print("Password Set Successfully")
        break  # Exit loop when password is valid
```

```python
# Task 6 Display the Seven Times Table
for i in range(13):  # Numbers 0 to 12 inclusive
    result = 7 * i
    print(f"{i} x 7 = {result}")
```

```python
# Task 7 User enters which times table to display
table = int(input("Enter the number for the times table (0-12): "))

# Ensure the number is within the valid range
if 0 <= table <= 12:
    for i in range(13):
        print(f"{i} x {table} = {i * table}")
else:
    print("Error: Please enter a number between 0 and 12.")
```

```python
# Task 8 Times table with option to print backwards if negative
table = int(input("Enter the number for the times table (0-12, negative for backwards): "))

if abs(table) > 12:
    print("Error: Please enter a number between -12 and 12.")
else:
    if table >= 0:
        # Forward table
        for i in range(13):
            print(f"{i} x {table} = {i * table}")
    else:
        # Backwards table if negative
        for i in range(12, -1, -1):
            print(f"{i} x {abs(table)} = {i * abs(table)}")
```
