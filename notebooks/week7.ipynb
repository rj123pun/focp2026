```python
#Task1
def unique_sorted_letters(text):
    """
    This function takes a string and returns a sorted list
    of unique letters used in the string.
    """
    # Convert the string to lowercase so case does not matter
    text = text.lower()
    
    # Use a set to remove duplicate letters, then sort the result
    return sorted(set(text))


print(unique_sorted_letters("cheese"))
```

```python
#Task 2
def letters_in_either(word1, word2):
    # Letters that appear in at least one of the two words
    return sorted(set(word1.lower()) | set(word2.lower()))


def letters_in_both(word1, word2):
    # Letters that appear in both words
    return sorted(set(word1.lower()) & set(word2.lower()))


def letters_in_one_not_both(word1, word2):
    # Letters that appear in either word, but NOT in both
    return sorted(set(word1.lower()) ^ set(word2.lower()))

print(letters_in_either("hello", "world"))
print(letters_in_both("hello", "world"))
print(letters_in_one_not_both("hello", "world"))
```

```python
#Task3
# Dictionary to store countries and their capitals
countries = {}

while True:
    # Ask the user for a country name
    country = input("\nEnter a country name (or type 'quit' to exit): ").strip()
    
    # Allow the user to exit the program safely
    if country.lower() == "quit":
        print("Program ended. Goodbye!")
        break
    
    # Convert country name to lowercase for consistent storage
    country_key = country.lower()
    
    # Check if the country already exists in the dictionary
    if country_key in countries:
        print(f"The capital of {country.title()} is {countries[country_key]}.")
    else:
        # Ask for the capital if we don't already know it
        capital = input(f"I don't know the capital of {country.title()}. Please enter it: ")
        countries[country_key] = capital
        print("Thank you! I've saved that information.")
```

```python
#task 4 Ask the user for an encrypted or normal message
message = input("Enter a message: ").lower()

# Dictionary to store letter counts
letter_count = {}

# Count each letter in the message
for char in message:
    # Only count alphabet letters (ignore spaces, numbers, symbols)
    if char.isalpha():
        # If letter not seen before, start count at 0
        if char not in letter_count:
            letter_count[char] = 0
        
        # Increase the count for this letter
        letter_count[char] += 1

# Convert dictionary into a list of (letter, count) pairs
letter_list = list(letter_count.items())

# Sort letters by count (highest first)
letter_list.sort(key=lambda item: item[1], reverse=True)

# Display the top 6 most common letters
print("\nSix most common letters:")
for letter, count in letter_list[:6]:
    print(f"{letter} : {count}")
```
