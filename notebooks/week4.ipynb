```python
# task 1 checking This function whether a number is between 0 and 100 
def is_valid_range(number):
    return 0 <= number <= 100

# testing
user_input = int(input("Enter an integer: "))

if is_valid_range(user_input):
    print("The number is within the valid range (0–100).")
else:
    print("The number is outside the valid range.")
```

```python
# task 2 checking wheather This function counts uppercase and lowercase letters in a string
def count_case_letters(text):
    uppercase_count = 0
    lowercase_count = 0

    for char in text:
        if char.isupper():
            uppercase_count += 1
        elif char.islower():
            lowercase_count += 1

    return uppercase_count, lowercase_count


# Testing
user_text = input("Enter a string: ")
upper, lower = count_case_letters(user_text)

print("Uppercase letters:", upper)
print("Lowercase letters:", lower)
```

```python
# Task 4 This function removes the last character of a string
def remove_last_character(text):
    if len(text) <= 1:
        return text  # Return unchanged if string is too short
    return text[:-1]


# Testing Program 
user_text = input("Enter some text: ")
result = remove_last_character(user_text)

print("Result:", result)
```

```python
#Task 5 Converting Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9 / 5) + 32


# Converts Fahrenheit to Celsius
def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5 / 9


#  Test Program 
c = float(input("Enter temperature in Celsius: "))
print("Fahrenheit:", celsius_to_fahrenheit(c))

f = float(input("Enter temperature in Fahrenheit: "))
print("Celsius:", fahrenheit_to_celsius(f))
```

```python
# Task 6 Reading temperature in the format "25C"
temp_input = input("Enter temperature (e.g. 25C): ")

# Extract number and convert it
celsius_value = float(temp_input[:-1])
fahrenheit_value = (celsius_value * 9 / 5) + 32

print(f"{fahrenheit_value:.1f}F")
```

```python
#task 7 
import statistics

temperatures = []

# Read exactly 6 temperatures
for i in range(6):
    temp_input = input(f"Enter temperature #{i + 1} (e.g. 20C): ")
    celsius_value = float(temp_input[:-1])
    temperatures.append(celsius_value)

print("Maximum:", max(temperatures))
print("Minimum:", min(temperatures))
print("Mean:", statistics.mean(temperatures))
```

```python
#task 7
import statistics

temperatures = []

while True:
    temp_input = input("Enter temperature (or press Enter to finish): ")

    if temp_input == "":
        break  # Stop when user presses Enter

    celsius_value = float(temp_input[:-1])
    temperatures.append(celsius_value)

# Only calculate if values were entered
if temperatures:
    print("Maximum:", max(temperatures))
    print("Minimum:", min(temperatures))
    print("Mean:", statistics.mean(temperatures))
else:
    print("No temperatures were entered.")
```
