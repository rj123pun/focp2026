```python

''' Task 2: 
 This program converts a temperature in Celsius to Fahrenheit.'''


# Prompt the user to enter a temperature in Celsius
celsius = float(input("\nEnter a temperature in Celsius: "))

# Convert Celsius to Fahrenheit using the formula F = C * 9/5 + 32
fahrenheit = (celsius * 9 / 5) + 32

# Display the result
print(f"{celsius}C is equivalent to {fahrenheit}F.")
```

```python
''' Task 3 
 This program calculates how many lab groups are needed and
 how many students will be left over.'''


# Ask the user for number of students and desired group size
total_students = int(input("\nHow many students? "))
group_size = int(input("Required group size? "))

# Calculate number of full groups
num_groups = total_students // group_size

# Calculate leftover students
leftover_students = total_students % group_size

# Handle plural/singular in the output for better grammar
student_word = "student" if leftover_students == 1 else "students"

# Display the result
print(f"There will be {num_groups} groups with {leftover_students} {student_word} left over.")
```

```python
''' Task 4
 This program calculates how many sweets each pupil gets
 and how many are left over.'''

# Ask the teacher for total sweets and number of pupils
total_sweets = int(input("\nHow many sweets are in the tub? "))
num_pupils = int(input("How many pupils today? "))

# Calculate sweets per pupil and leftover sweets
sweets_per_pupil = total_sweets // num_pupils
leftover_sweets = total_sweets % num_pupils

# Handle plural/singular for better grammar
sweets_word = "sweet" if sweets_per_pupil == 1 else "sweets"
leftover_word = "sweet" if leftover_sweets == 1 else "sweets"

# Display the result
print(f"Each pupil gets {sweets_per_pupil} {sweets_word}.")
print(f"The teacher will have {leftover_sweets} {leftover_word} left over.")
```
