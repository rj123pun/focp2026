```python
#task 1. Convert a positive integer to binary

def to_binary(number):
    """
    This function takes a positive integer
    and returns its binary (base 2) representation.
    Python already provides a built-in way to do this.
    """
    if number <= 0:
        return "Please enter a positive integer"

    # bin() converts a number to binary but adds '0b' at the start,
    # so we remove it using slicing.
    return bin(number)[2:]

print("1. Binary Conversion")
print("10 in binary:", to_binary(10))
print()
```

```python
#Task 2 Find all factors of an integer

def find_factors(number):
    """
    This function finds and returns all factors of a given integer.
    A factor divides the number exactly with no remainder.
    """
    factors = []

    for i in range(1, number + 1):
        if number % i == 0:
            factors.append(i)

    return factors

print("2. Factors")
print("Factors of 12:", find_factors(12))
print()
```

```python
# Task3 Check if a number is prime
# -------------------------------------------------
def is_prime(number):
    """
    This function checks whether a number is prime.
    A prime number is greater than 1 and has only two factors.
    """
    if number <= 1:
        return False

    for i in range(2, number):
        if number % i == 0:
            return False

    return True

print("3. Prime Check")
print("Is 13 prime?", is_prime(13))
print("Is 20 prime?", is_prime(20))
print()

```

```python
# Task 5 Hide a message using a random interval

def hide_message(message):
    """
    This function hides a message inside random letters.
    The real letters appear at a random interval (2–20).
    It returns the encrypted message and the interval used.
    """
    interval = random.randint(2, 20)
    encrypted = ""
    alphabet = string.ascii_lowercase

    for letter in message.replace(" ", ""):
        encrypted += letter
        for _ in range(interval - 1):
            encrypted += random.choice(alphabet)

    return encrypted, interval

print("5. Hide Message")
secret, step = hide_message("send cheese")
print("Encrypted message:", secret)
print("Interval used:", step)
print() 
```

```python
#Task 6 Decrypt a hidden message

def decrypt_message(encrypted_message, interval):
    """
    This function extracts the original message
    from an encrypted string using the known interval.
    """
    decrypted = ""

    for i in range(0, len(encrypted_message), interval):
        decrypted += encrypted_message[i]

    return decrypted

print("6. Decrypt Message")
print("Decrypted message:", decrypt_message(secret, step))
```

```python
#Task 4
def simple_encrypt(message):
    """
    This function removes spaces from a message
    and then reverses the string to 'encrypt' it.
    """
    no_spaces = message.replace(" ", "")
    encrypted = no_spaces[::-1]  # reverse the string
    return encrypted

print("4. Simple Encryption")
message = "hello world"
print("Original:", message)
print("Encrypted:", simple_encrypt(message))
print()
```
