```python
#task1
''' This program shows which operating system is being used.It uses the built-in sys module to get platform information.'''

import sys

# sys.platform gives a short name for the OS
print("You are running this program on:")
print(sys.platform)
```

```python
# This program counts how many command-line arguments were provided.
# The program name itself is NOT counted as an argument.

import sys

# Subtract 1 because the first item is the program name
argument_count = len(sys.argv) - 1

print("Number of arguments provided:", argument_count)
```

```python
# This program finds the shortest command-line argument.
# If multiple arguments have the same shortest length, any one is fine.

import sys

# Remove the program name from the argument list
arguments = sys.argv[1:]

# Check if the user provided any arguments
if not arguments:
    print("Error: No arguments provided.")
    sys.exit()

# Sort arguments by length (shortest first)
arguments.sort(key=len)

# The first item will now be the shortest
print("Shortest argument:", arguments[0])
```

```python
# This program checks if a website is reachable.
# It does this by looking at the HTTP response code.

import sys
import urllib.request

# Check if a URL was provided
if len(sys.argv) < 2:
    print("Error: Please provide a URL.")
    sys.exit()

url = sys.argv[1]

try:
    # Try to open the URL
    response = urllib.request.urlopen(url)
    print("Website is working!")
    print("HTTP Status Code:", response.getcode())

except:
    # If anything goes wrong, the website is not reachable
    print("Website is not reachable or URL is invalid.")
```

```python
# This program calculates the minimum, maximum, and average temperature
# using values provided through the command line.

import sys

# Get temperature values (excluding program name)
arguments = sys.argv[1:]

# Check if any temperatures were provided
if not arguments:
    print("Error: No temperature values provided.")
    sys.exit()

# Convert arguments to floats
temperatures = []
for value in arguments:
    temperatures.append(float(value))

# Calculate statistics
minimum = min(temperatures)
maximum = max(temperatures)
average = sum(temperatures) / len(temperatures)

# Display results
print("Minimum temperature:", minimum)
print("Maximum temperature:", maximum)
print("Average temperature:", round(average, 2))
```

```python
# This program creates a backup copy of a given file.
# The backup will have the same contents but a different name.

import sys
import shutil

# Check if a file name was provided
if len(sys.argv) < 2:
    print("Error: Please provide a file name.")
    sys.exit()

original_file = sys.argv[1]
backup_file = original_file + ".bak"

try:
    # Copy the file contents to a new backup file
    shutil.copy(original_file, backup_file)
    print("Backup created successfully:", backup_file)

except:
    # Handle cases where the file does not exist or cannot be copied
    print("Error: File could not be backed up.")
```
